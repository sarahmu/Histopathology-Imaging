```python
from PIL import Image
import numpy as np
import os
```

```python
orginal_dir = 'data/good_patches/'
gray_dir = 'data/good_patches_gray/'
```

```python
counter = 0
for file in os.listdir(orginal_dir):
    counter += 1
    if counter % 100 == 0:
        print(counter)
    if file.endswith('.png'):
        img = Image.open(orginal_dir + file)
        gray_img = img.convert('L')
        gray_img.save(gray_dir + 'gray_' + file)
```
